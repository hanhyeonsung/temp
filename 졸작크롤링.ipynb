```python

downloads_path = os.path.expanduser("~/Downloads")
csv_files = glob.glob(os.path.join(downloads_path, "*.csv"))

today = datetime.today().date()
today_files = []

for file in csv_files:
    modified_time = datetime.fromtimestamp(os.path.getmtime(file)).date()
    if modified_time == today:
        today_files.append(file)

```

```python
from io import StringIO
import pandas as pd 

temp_df = pd.DataFrame([])

for i in today_files: 
    with open(i, 'rb') as f:
        lines = f.readlines()

    # bytes 처리
    valid_lines = [line for line in lines if not line.strip().startswith(b'{')]

    # bytes → str 디코딩 후 DataFrame으로 변환
    decoded_text = b''.join(valid_lines).decode('cp949')  # 또는 'utf-8'
    df = pd.read_csv(StringIO(decoded_text))
    temp_df = pd.concat([temp_df, df])
temp_df.shape 
```
